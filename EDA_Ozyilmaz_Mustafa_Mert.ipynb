```python

import pandas as pd

# Load the data from the uploaded CSV files
cab_data = pd.read_csv(r'C:\Users\MUSTAFA MERT\Desktop\dataGlacier week2\DataSets-main\Cab_Data.csv')
customer_data = pd.read_csv(r'C:\Users\MUSTAFA MERT\Desktop\dataGlacier week2\DataSets-main\Customer_ID.csv')
city_data = pd.read_csv(r'C:\Users\MUSTAFA MERT\Desktop\dataGlacier week2\DataSets-main\City.csv')
transaction_data = pd.read_csv(r'C:\Users\MUSTAFA MERT\Desktop\dataGlacier week2\DataSets-main\Transaction_ID.csv')

# Display the first few rows of each dataset to understand their structure
cab_data_head = cab_data.head()
customer_data_head = customer_data.head()
city_data_head = city_data.head()
transaction_data_head = transaction_data.head()

cab_data_head, customer_data_head, city_data_head, transaction_data_head




```

```python
# Data Cleaning

# Cab Data: Checking for missing values and data types
cab_data_info = cab_data.info()
cab_data_missing = cab_data.isnull().sum()

# Customer Data: Checking for missing values and data types
customer_data_info = customer_data.info()
customer_data_missing = customer_data.isnull().sum()

# City Data: Checking for missing values and data types
city_data_info = city_data.info()
city_data_missing = city_data.isnull().sum()

# Transaction Data: Checking for missing values and data types
transaction_data_info = transaction_data.info()
transaction_data_missing = transaction_data.isnull().sum()

# Display the information and missing values summary
cab_data_info, cab_data_missing, customer_data_info, customer_data_missing, city_data_info, city_data_missing, transaction_data_info, transaction_data_missing
```

```python
# Step 1: Standardize Date Format in Cab Data
cab_data['Date of Travel'] = pd.to_datetime(cab_data['Date of Travel'], origin='1899-12-30', unit='D')

# Step 2: Correct Data Types in City Data
# Remove commas and convert to int
city_data['Population'] = city_data['Population'].str.replace(',', '').astype(int)
city_data['Users'] = city_data['Users'].str.replace(',', '').astype(int)

# Step 3: Check for Duplicates
cab_data_duplicates = cab_data.duplicated().sum()
customer_data_duplicates = customer_data.duplicated().sum()
city_data_duplicates = city_data.duplicated().sum()
transaction_data_duplicates = transaction_data.duplicated().sum()

# Step 4: Handle Outliers
# Identify outliers using IQR for numeric columns in cab_data
Q1 = cab_data[['KM Travelled', 'Price Charged', 'Cost of Trip']].quantile(0.25)
Q3 = cab_data[['KM Travelled', 'Price Charged', 'Cost of Trip']].quantile(0.75)
IQR = Q3 - Q1
outliers = ((cab_data[['KM Travelled', 'Price Charged', 'Cost of Trip']] < (Q1 - 1.5 * IQR)) | (cab_data[['KM Travelled', 'Price Charged', 'Cost of Trip']] > (Q3 + 1.5 * IQR))).sum()

# Display results
cab_data.head(), city_data.head(), cab_data_duplicates, customer_data_duplicates, city_data_duplicates, transaction_data_duplicates, outliers
```

```python
# Handling Outliers by capping values at the 1.5 * IQR limits
Q1 = cab_data['Price Charged'].quantile(0.25)
Q3 = cab_data['Price Charged'].quantile(0.75)
IQR = Q3 - Q1
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

cab_data['Price Charged'] = cab_data['Price Charged'].apply(lambda x: upper_limit if x > upper_limit else (lower_limit if x < lower_limit else x))

# Data Integration: Join cab_data with transaction_data on 'Transaction ID'
master_data = pd.merge(cab_data, transaction_data, on='Transaction ID', how='left')

# Join the resulting dataframe with customer_data on 'Customer ID'
master_data = pd.merge(master_data, customer_data, on='Customer ID', how='left')

# Join the resulting dataframe with city_data on 'City'
master_data = pd.merge(master_data, city_data, on='City', how='left')

# Display the first few rows of the master dataset
master_data_head = master_data.head()
master_data_head
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
from scipy import stats

# Hypothesis 1: Seasonality in Cab Usage
seasonality_test = stats.chisquare(monthly_trips)

# Hypothesis 2: Revenue and Distance Relationship
correlation_test = stats.pearsonr(master_data['KM Travelled'], master_data['Price Charged'])

# Hypothesis 3: Company Performance
pink_cab_revenue = master_data[master_data['Company'] == 'Pink Cab']['Price Charged']
other_cab_revenue = master_data[master_data['Company'] != 'Pink Cab']['Price Charged']
company_performance_test = stats.ttest_ind(pink_cab_revenue, other_cab_revenue, equal_var=False)

# Hypothesis 4: Customer Demographics Impact (Age and Income)
# Multiple linear regression
X = master_data[['Age', 'Income (USD/Month)']]
y = master_data['KM Travelled']
X = sm.add_constant(X)
age_income_regression = sm.OLS(y, X).fit()

# Hypothesis 5: City-wise Cab Usage
city_usage_test = stats.f_oneway(*[group['KM Travelled'].values for name, group in master_data.groupby('City')])

# Hypothesis 6: Payment Mode Preference
payment_mode_counts = master_data['Payment_Mode'].value_counts()
payment_mode_test = stats.chisquare(payment_mode_counts)

# Displaying the test results
{
    "Seasonality Test": seasonality_test,
    "Correlation Test": correlation_test,
    "Company Performance Test": company_performance_test,
    "Age and Income Regression Summary": age_income_regression.summary(),
    "City Usage Test": city_usage_test,
    "Payment Mode Preference Test": payment_mode_test
}
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import numpy as np

# Aggregate customer data
customer_agg = master_data.groupby('Customer ID').agg({
    'KM Travelled': 'sum',
    'Price Charged': 'sum',
    'Age': 'first',
    'Income (USD/Month)': 'first'
}).reset_index()

# Rename columns for clarity
customer_agg.columns = ['Customer ID', 'Total KM Travelled', 'Total Price Charged', 'Age', 'Income']

# Data normalization
scaler = StandardScaler()
customer_agg_scaled = scaler.fit_transform(customer_agg[['Total KM Travelled', 'Total Price Charged', 'Age', 'Income']])

# Determine the optimal number of clusters using the Elbow method
sse = {}
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(customer_agg_scaled)
    sse[k] = kmeans.inertia_

# Plot the Elbow method graph
plt.figure(figsize=(10, 6))
plt.plot(list(sse.keys()), list(sse.values()), marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.savefig("7")
plt.show()

# Choose an optimal number of clusters (let's assume 4 based on the Elbow method) and perform K-means clustering
optimal_clusters = 4
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
customer_agg['Cluster'] = kmeans.fit_predict(customer_agg_scaled)

# Analyze the characteristics of each cluster
cluster_analysis = customer_agg.groupby('Cluster').mean().reset_index()



cluster_analysis
```

```python
from sklearn.decomposition import PCA

# Perform PCA to reduce dimensions to 2 for visualization
pca = PCA(n_components=2)
principal_components = pca.fit_transform(customer_agg_scaled)
customer_agg['PCA1'] = principal_components[:, 0]
customer_agg['PCA2'] = principal_components[:, 1]

# Visualize the clusters
plt.figure(figsize=(12, 8))
sns.scatterplot(x='PCA1', y='PCA2', hue='Cluster', data=customer_agg, palette='viridis', s=100, alpha=0.7)
plt.title('Customer Segmentation Clusters')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Cluster')
plt.show()



```

## Hypothesis Testing Results

### Seasonality in Cab Usage:
- **Chi-square Test:** \(\chi^2 = 33943.20\), \(p\)-value = 0.0
- **Result:** There is a significant variation in cab usage across different months, indicating seasonality.

### Revenue and Distance Relationship:
- **Pearson Correlation Test:** \(r = 0.8455\), \(p\)-value = 0.0
- **Result:** There is a strong positive correlation between the distance travelled and the revenue generated.

### Company Performance:
- **Independent T-Test:** \(t = -175.73\), \(p\)-value = 0.0
- **Result:** There is a significant difference in revenue generated by the two companies, with one company consistently generating higher revenue.

### Customer Demographics Impact:
- **Multiple Linear Regression:**
  - Age Coefficient: \(-0.0004\), \(p\)-value = 0.826
  - Income Coefficient: \(-8.33 \times 10^{-7}\), \(p\)-value = 0.745
- **Result:** Age and income do not have a significant impact on the distance travelled.

### City-wise Cab Usage:
- **ANOVA Test:** \(F = 1.196\), \(p\)-value = 0.254
- **Result:** There are no significant differences in cab usage across different cities.

### Payment Mode Preference:
- **Chi-square Test:** \(\chi^2 = 14270.91\), \(p\)-value = 0.0
- **Result:** There is a significant preference for one payment mode over another.
